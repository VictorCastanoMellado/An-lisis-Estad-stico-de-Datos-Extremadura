### 2. ACTIVIDAD PRODUCTIVA

### Variación Producto Interior Bruto (Volumen)

```python
pibesp=pd.read_excel(excel,sheet_name="PIB Spain", index_col=0, parse_dates=True)
pibext=pd.read_excel(excel, sheet_name="PIB Extremadura", index_col=0, parse_dates=True)
desc_pibesp=sm.tsa.x13_arima_analysis(endog = pibesp,x12path=XPATH,trading=True)
desc_pibext=sm.tsa.x13_arima_analysis(endog = pibext,x12path=XPATH,trading=True)
tvpibesp=desc_pibesp.seasadj.diff()/desc_pibesp.seasadj.shift()*100
tvpibext=desc_pibext.seasadj.diff()/desc_pibext.seasadj.shift()*100

fig, ax=plt.subplots(figsize=(10,3))
ax.plot(tvpibesp, color='green', label='PIB Extremadura')
ax.plot(tvpibext, color='red', label='PIB España')
ax.set_ylabel("Variación")
ax.set_xlabel("Años")
ax.legend()
ax.set_title("Variación PIB", size=14, pad=10)
fig.text(0.25,-0.03,"Serie desestacionalizada. Elaboración propia. Fuente: INE.", ha="center",fontsize=8 )
plt.show()
```

<hr>El PIB de España disminuye alrededor de finales de 2019, hasta un –10% y a partir de ahí, se recupera y se mantiene estable desde el inicio de 2020.
El PIB en Extremadura baja en el 2020 llegando incluso por debajo del -10% y a partir de ahí sufre un periodo de altibajos hasta que se mantiene estable desde el 2º trimestre de 2021.

### Estimación Afiliados a la Seguridad sin COVID-19 Extremadura

```python
ss=ssext[:51]
sscovid=ssext[50:]
ssnocovid = sm.tsa.statespace.SARIMAX(ss,order=(2,1,2),seasonal_order=(2,0,1,4), trend="c",
                                    enforce_stationarity=False, enforce_invertibility=False)
predssnocovid=ssnocovid.fit()
predssnocovid.summary
tp_ssnocovid = predssnocovid.get_prediction(start='2020T3',end='2022T12', freq='T')
index_of_fc5 = pd.date_range(start="2020T3",end="2022T12", freq='T')
fitted_series5 = pd.Series(tp_ssnocovid.predicted_mean, index=index_of_fc5)

fig, ax=plt.subplots(figsize=(10,5))
ax.plot(tp_ssnocovid.predicted_mean, color='orange', label='Predicción sin crisis del COVID-19')
ax.plot(ss, color='green', label='Afiliados a la Seguridad Social')
ax.plot(sscovid, color='blue', label='Tasa de desempleados durante la crisis del COVID-19')
ax.set_ylabel('Afiliados Seguridad Social')
ax.set_xlabel('Años')
ax.legend()
ax.set_title('Afiliados Seguridad Social: Situacion COVID-19 vs Predicción sin COVID-19', size=14, pad=10)
fig.text(0.25,-0.03,"Serie Original. Elaboración propia. Fuente: Ministerio de Inclusión,Seguridad Social y Migraciones", ha="center",fontsize=8)
plt.show()
```

<hr>
En cuanto a los afiliados de la seguridad social, estos van aumentando siguiendo una tendencia alcista y aparentemente estacional hasta el 2º trimestre de 2019, momento en el que la serie empieza a descender llegando al 2º trimestre de 2020 en valores casi de 2017.

En cambio, en la serie de afiliados durante la crisis, vemos como en contrario a la hipótesis anterior la serie va en aumento hasta llegar a niveles máximos en el 2º trimestre de 2022.


### Estimación Tasa de Paro sin COVID-19 Extremadura
